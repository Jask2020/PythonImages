```python
dog = cv.imread('dog.jpg')
gray = cv.cvtColor(dog, cv.COLOR_BGR2GRAY)
```

```python
grayimg = Image.fromarray(gray)
```

```python
display(grayimg)
```

```python
print(cv2.__file__)
```

```python
faceCascade = cv.CascadeClassifier('cvdata/haarcascade_frontalface_alt.xml')
eyeCascade = cv.CascadeClassifier('cvdata/haarcascade_eye_tree_eyeglasses.xml')
```

```python
img = cv.imread('faceimg.jpg')
grayimg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
display(Image.fromarray(grayimg))
```

```python
faces = faceCascade.detectMultiScale(grayimg)
eyes = eyeCascade.detectMultiScale(grayimg)
```

```python
faces
```

```python
eyes
eye1 = eyes.tolist()[0]
eye2 = eyes.tolist()[1]

print(eye1)
```

```python
pilImg = Image.fromarray(grayimg)
draw = ImageDraw.Draw(pilImg)
draw.rectangle((eye1[0],eye1[1],eye1[0]+eye1[2],eye1[1]+eye1[3]), fill=None, outline='white')
draw.rectangle((eye2[0],eye2[1],eye2[0]+eye2[2],eye2[1]+eye2[3]), fill=None, outline='white')
```

```python
display(pilImg)

print(Image.open('faceimg.jpg').mode)
print(pilImg.mode)
```
